```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Update the path below to the actual location of your headbrain.csv file
data = pd.read_csv('/content/headbrain.csv')
data.head()
```

```python
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
mean_x = np.mean(X)
mean_y = np.mean(Y)
n = len(X)
```

```python
numerator = np.sum((X - mean_x) * (Y - mean_y))
denominator = np.sum((X - mean_x) ** 2)
w1 = numerator / denominator
w0 = mean_y - (w1 * mean_x)

print(f"Manual Linear Regression Coefficients: w0 = {w0:.2f}, w1 = {w1:.2f}")
```

```python
def predict(x):
    return w0 + w1 * x
```

```python
Y_pred = predict(X)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel("Head Size (cm³)")
plt.ylabel("Brain Weight (grams)")
plt.title("Linear Regression using Least Squares Method")
plt.legend()
plt.grid(True)
plt.show()
```

```python
mse_manual = mean_squared_error(Y, Y_pred)
print(f"Manual Model MSE: {mse_manual:.2f}")
```

```python
X_reshaped = X.reshape(-1, 1)
reg = LinearRegression()
reg.fit(X_reshaped, Y)
Y_pred_sklearn = reg.predict(X_reshaped)

print(f"Scikit-learn Coefficients: w0 = {reg.intercept_:.2f}, w1 = {reg.coef_[0]:.2f}")
mse_sklearn = mean_squared_error(Y, Y_pred_sklearn)
print(f"Scikit-learn Model MSE: {mse_sklearn:.2f}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred_sklearn, color='green', label='Sklearn Regression Line')
plt.xlabel("Head Size (cm³)")
plt.ylabel("Brain Weight (grams)")
plt.title("Linear Regression using scikit-learn")
plt.legend()
plt.grid(True)
plt.show()
```
