```python
import matplotlib.pyplot as plt
import seaborn as sns

# Corrected file path
file_path = '/content/drive/MyDrive/IRIS.csv'
df = pd.read_csv(file_path)

print(df.head())

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plt.figure(figsize=(12, 8))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=df[feature], y=[0]*len(df), hue=df['species'])
    plt.xlabel(feature)
    plt.yticks([])
    plt.title(f'Univariate Scatter Plot of {feature}')
plt.tight_layout()
plt.show()
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x='sepal_width', y='petal_width', hue='species', data=df)
plt.title('sepal_width vs petal_width')

plt.subplot(1, 2, 2)
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title('sepal_length vs petal_length')

plt.tight_layout()
plt.show()

if 'Id' in df.columns:
    df_features = df.drop(columns=['Id'])
else:
    df_features = df.copy()

sns.pairplot(df_features, hue='species', diag_kind='kde')
plt.suptitle('Pairplot of Iris Features with KDE', y=1.02)
plt.show()
```
